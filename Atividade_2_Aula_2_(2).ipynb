Codificação de variveis categoricas

```python
#Codificação de variaveis categoricas (que tem strings)

#

from sklearn.preprocessing import LabelEncoder

#lsita com variveis categoricas
categorical_cols = ['person_home_ownership', 'loan_intent', 'loan_grade', 'cb_person_default_on_file']


#lable vem da biblioteca do sklearn.preprocessing  que foi importado
#que vai fazer um 'le.fit_transform' ou seja, transformar as strings em valoeres numericos para poder manipular
#portanto as lista das colunas serao trasnormads em valoers nuemricos--> 'person_home_ownership', 'loan_intent', 'loan_grade', 'cb_person_default_on_file'
le = LabelEncoder()
for col in categorical_cols:
    base_credit[col] = le.fit_transform(base_credit[col])
```

##Outliers

Os Outliers --> ponttos fora do padrao que podem prejudicar o dataset
INTERQUATILE RANGE (IQR)

```python
for col in ['person_income', 'loan_amnt']:
  Q1 = base_credit[col].quantile(0.25)
  Q3 = base_credit[col].quantile(0.75)
  IQR = Q3 - Q1
  lower_bound = Q1 - 1.5 * IQR
  upper_bound = Q3 + 1.5 * IQR
  base_credit = base_credit[(base_credit[col] >= lower_bound) & (base_credit[col] <= upper_bound)]

  base_credit.head()
```
